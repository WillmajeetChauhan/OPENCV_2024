```python
import cv2
import numpy as np
```

```python
# corner detection (harris corner detection)
img = cv2.imread(r"C:\Users\HP LAPTOP-PC\Desktop\DRDO Project\images.jpeg")
gr = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#print(gr)

gr = np.float32(gr)  #convert int matrix into foat matrix of img.

res = cv2.cornerHarris(gr,5,3,0.04)# detect the corner of img
#src, blockSize, ksize, k[, dst[, borderType]])

res = cv2.dilate(res,None)  # to remove irrelevant corner point.
#print(res)
img[res>0.01*res.max()]= [0,0,255]  # to impose corner detection on org img



cv2.imshow("willma",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#C D (USING shi tomasi corner detection) we can choose how many corner points we want to identify.  

img = cv2.imread(r"C:\Users\HP LAPTOP-PC\Desktop\DRDO Project\images.jpeg")
img = cv2.resize(img,(500,500))
gr = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

cr = cv2.goodFeaturesToTrack(gr,35,0.01,20)  # detect corner points
#(image, maxCorners, qualityLevel, minDistance)
#print(cr)

cr = np.int64(cr)  # convert float values into int type.
for i in cr:
    x,y = i.ravel()  # extract coordinates from cr and convert them in 1 dimension.
    cv2.circle(img,(x,y),5,(0,0,255),-1)
    #(img, center, radius, color[, thickness])


cv2.imshow("willma",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# object detection using haar-cascade(face detection) haar cascade file used for this.

img = cv2.imread("IMG100.jpeg")
img = cv2.resize(img,(600,500))

gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

f = cv2.CascadeClassifier(r"C:\Users\HP LAPTOP-PC\Desktop\DRDO Project\haarcascade_frontalface_default.xml") # to get the haar cascade file.
d = f.detectMultiScale(gry,1.3,5)  #extract our coordinates from image to detect face.

#print(d)

for (x,y,w,h) in d:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),3)
    #(img, pt1, pt2, color[, thickness[, lineType[, shift]]])

cv2.imshow("willma1",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# displaying coordinates and color code of the points clicked on the image. 
def click_b(event,x,y,f,p):  # which click on mouse performs which action. coordinate,flag = empty,parameter.
    if event == cv2.EVENT_LBUTTONDOWN:
        s = f"{x},{y}"
        cv2.putText(img,s,(x,y),cv2.FONT_HERSHEY_PLAIN,0.5,(0,0,0))
#(img, text, org, fontFace, fontScale, color)
        cv2.imshow("willma",img)
    elif event == cv2.EVENT_RBUTTONDOWN:
        b = img[y,x,0]
        g = img[y,x,1]
        r = img[y,x,2]
        s = f"{b},{g},{r}"
        cv2.putText(img,s,(x,y),cv2.FONT_HERSHEY_COMPLEX,0.5,(0,0,255))
        cv2.imshow("willma",img)

img = cv2.imread("img103.jpeg")

cv2.imshow("willma",img)

cv2.setMouseCallback("willma",click_b)# (windowName, onMouse)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# smile  detection.
img = cv2.imread("img3.jpg")
gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

sm = cv2.CascadeClassifier(r"C:\Users\HP LAPTOP-PC\Desktop\DRDO Project\haarcascade_smile.xml")
# firstly separate the region of interest,so that it does not detect any wrong smile.
fc = cv2.CascadeClassifier(r"C:\Users\HP LAPTOP-PC\Desktop\DRDO Project\haarcascade_frontalface_default.xml")

f = fc.detectMultiScale(gry,1.2,3)

for x,y,w,h in f:
    roi_gry = gry[y:y+h,x:x+w]
    roi_img = img[y:y+h,x:x+w]

    s = sm.detectMultiScale(roi_img,1.5,9)

    for xs,ys,ws,hs,in s:
        cv2.rectangle(roi_img,(xs,ys),(xs+ws,ys+hs),(0,255,0),2)
        
    

cv2.imshow("willmajeet",img)
#cv2.imshow("willma",roi_img)
#cv2.imshow("WILLMA",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
