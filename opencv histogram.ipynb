```python
img = cv2.imread("img2.jpg")
w,h = (667, 1000)

src1 = np.float32([[164,99],[172,475],[967,88],[964,479]])
dst1 = np.float32([[0,0],[w,0],[0,h],[w,h]])
m = cv2.getPerspectiveTransform(src1,dst1)  # perfomed in an tillted image.
new_img = cv2.warpPerspective(img,m,(w,h))

cv2.imshow("willma",new_img)
cv2.imshow("willmajeet",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# threshold image
img = cv2.imread("img12.jpg")
img = cv2.resize(img,(600,400))

_,th = cv2.threshold(img,190,255,cv2.THRESH_BINARY)  # 0-255 image can vary.
cv2.imshow("THRESHOLD",th)
cv2.imshow("willma",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# adaptive histogram equilizer 

img = cv2.imread("img30.jpg")
img = cv2.resize(img,(500,500))

gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([gry],[0],None,[255],[0,255])
eq = cv2.equalizeHist(gry)

plt.plot(eq,c = "red")
plt.plot(hist)
plt.show()

cv2.imshow("willma1",gry)
cv2.imshow("willma",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# clahe histogram equilizer ( clear that image not done by adaptive equlizer)

img = cv2.imread("img31.jpg")
img = cv2.resize(img,(500,500))

gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([gry],[0],None,[255],[0,255])

eq = cv2.equalizeHist(gry)
hist1 = cv2.calcHist([eq],[0],None,[255],[0,255])

cl = cv2.createCLAHE(clipLimit = 1, tileGridSize = (8,8))
f = cl.apply(gry)

plt.plot(hist)
plt.plot(hist1 ,c = "red")
plt.show()

cv2.imshow("willma2",f)
cv2.imshow("willma1",eq)
cv2.imshow("willma",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
