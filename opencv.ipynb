```python
# color picker

img =np.zeros((500,500,3),np.uint8)*255
cv2.namedWindow("colour")  # add window  over image.

cv2.createTrackbar("R","colour",0,255,willma)
cv2.createTrackbar("G","colour",0,255,willma)
cv2.createTrackbar("B","colour",0,255,willma)

while True:
    cv2.imshow("colour",img)
    if cv2.waitKey(1) & 0xff == ord("q"):
        break
    r = cv2.getTrackbarPos("R","colour") # to get different colour to the trackbar.
    g = cv2.getTrackbarPos("G","colour")
    b = cv2.getTrackbarPos("B","colour")

    img[:] =[b,g,r]  # to bgr from rgb


cv2.destroyAllWindows()
```

```python
# filter colour with opencv( live object filter)
cap = cv2.VideoCapture(0)
def willma(x):
    pass

cv2.namedWindow("demo")
cv2.createTrackbar("lb","demo",0,255,willma)
cv2.createTrackbar("lg","demo",0,255,willma)
cv2.createTrackbar("lr","demo",0,255,willma)

cv2.createTrackbar("ub","demo",255,255,willma)
cv2.createTrackbar("ug","demo",255,255,willma)
cv2.createTrackbar("ur","demo",255,255,willma)

while cap.isOpened():
        r,frame = cap.read()

        if r == True:
                img = cv2.resize(frame,(400,300))
                hsv_img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

                Lb = cv2.getTrackbarPos("lb","demo")
                Lg = cv2.getTrackbarPos("lg","demo")
                Lr = cv2.getTrackbarPos("lr","demo")

                Ub = cv2.getTrackbarPos("ub","demo")
                Ug = cv2.getTrackbarPos("ug","demo")
                Ur = cv2.getTrackbarPos("ur","demo")

                lo = np.array([Lb,Lg,Lr])
                up = np.array([Ub,Ug,Ur])

                masks = cv2.inRange(hsv_img,lo,up)

                res = cv2.bitwise_and(img,img,mask = masks)

                cv2.imshow("res",res)
                cv2.imshow("mask",masks)
                cv2.imshow("hsv",hsv_img)
                cv2.imshow("willma",img)
                if cv2.waitKey(1)  & 0xff == ord("q"):
                    break
        else:
            break
cap.release()
cv2.destroyAllWindows()








```

```python
# image contours using opencv ( build lines over images)

img = cv2.imread("Chrome-Logo.png")
img = cv2.resize(img,(600,500))

img1 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

r,thr = cv2.threshold(img1,240,255,cv2.THRESH_BINARY)
contour,hex = cv2.findContours(thr,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
print(contour) # points which make contour line  # contour index

new_img = cv2.drawContours(img,contour,-1,(255,0,0),2)
cv2.imshow("wc1",new_img)
# cv2.imshow("wc",thr)
# cv2.imshow("willmajeet",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
